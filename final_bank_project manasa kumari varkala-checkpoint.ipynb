```python
import random
import mysql.connector

# Database connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Royalassembly@1",
    database="banking_db"
)
c = conn.cursor()

# Create accounts table if not exists
c.execute('''CREATE TABLE IF NOT EXISTS accounts
             (account_number INT PRIMARY KEY, account_type VARCHAR(20), name VARCHAR(50),
             address VARCHAR(100), kyc VARCHAR(100), mobile VARCHAR(15), email VARCHAR(50), balance DECIMAL(10,2))''')


class Account:
    def __init__(self, acc_num, acc_type, name, address, kyc, mobile, email):
        self.acc_num = acc_num
        self.acc_type = acc_type
        self.name = name
        self.address = address
        self.kyc = kyc
        self.mobile = mobile
        self.email = email
        self.balance = 0.0

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            raise ValueError("Insufficient balance!")

    def check_balance(self):
        return self.balance

    def transfer(self, amount, recipient):
        if self.balance >= amount:
            self.balance -= amount
            recipient.deposit(amount)
            print("Transfer successful!")
        else:
            raise ValueError("Insufficient balance for transfer!")

    def save_account(self):
        c.execute("INSERT INTO accounts VALUES (%s, %s, %s, %s, %s, %s, %s, %s)",
                  (self.acc_num, self.acc_type, self.name, self.address, self.kyc, self.mobile, self.email, self.balance))
        conn.commit()

    def update_account(self):
        c.execute("UPDATE accounts SET balance=%s WHERE account_number=%s", (self.balance, self.acc_num))
        conn.commit()


class Bank:
    def __init__(self):
        self.accounts = []

    def create_account(self, acc_type, name, address, kyc, mobile, email):
        acc_num = random.randint(1000, 9999)
        account = Account(acc_num, acc_type, name, address, kyc, mobile, email)
        account.save_account()
        self.accounts.append(account)
        return account

    def get_account_by_number(self, acc_num):
        c.execute("SELECT * FROM accounts WHERE account_number=%s", (acc_num,))
        row = c.fetchone()
        if row:
            account = Account(row[0], row[1], row[2], row[3], row[4], row[5], row[6])
            account.balance = row[7]
            return account
        else:
            return None

    def get_total_accounts(self):
        c.execute("SELECT COUNT(*) FROM accounts")
        return c.fetchone()[0]


class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def change_password(self, new_password):
        self.password = new_password

    def authenticate(self, username, password):
        return self.username == username and self.password == password


class Admin(User):
    def __init__(self, username, password):
        super().__init__(username, password)

    def get_account_statistics(self):
        total_accounts = bank.get_total_accounts()
        print("Total Accounts:", total_accounts)


def create_tables():
    c.execute('''CREATE TABLE IF NOT EXISTS users
                 (username VARCHAR(20) PRIMARY KEY, password VARCHAR(20))''')

    # Check if admin user exists
    c.execute("SELECT * FROM users WHERE username='admin'")
    if not c.fetchone():
        create_admin_user()


def create_admin_user():
    print("\n==== Admin Account Creation ====")
    username = input("Enter username for admin account: ")
    password = input("Enter password for admin account: ")
    c.execute("INSERT INTO users VALUES (%s, %s)", (username, password))
    conn.commit()
    print("Admin account created successfully!")


def login():
    while True:
        username = input("Enter username: ")
        password = input("Enter password: ")

        if admin_user.authenticate(username, password):
            return admin_user
        elif user.authenticate(username, password):
            return user
        else:
            print("Invalid credentials! Please try again.")


def open_account():
    acc_type = input("Enter account type (savings/current/deposit): ")
    name = input("Enter your name: ")
    address = input("Enter your address: ")
    kyc = input("Enter KYC information: ")
    mobile = input("Enter your mobile number: ")
    email = input("Enter your email address: ")

    account = bank.create_account(acc_type, name, address, kyc, mobile, email)
    print("Account created successfully!")
    print("Account Number:", account.acc_num)


def deposit(account):
    amount = float(input("Enter the deposit amount: "))
    account.deposit(amount)
    account.update_account()
    print("Amount deposited successfully!")


def withdraw(account):
    amount = float(input("Enter the withdrawal amount: "))
    account.withdraw(amount)
    account.update_account()


def check_balance(account):
    balance = account.check_balance()
    print("Current Balance:", balance)


def transfer(account):
    recipient_num = input("Enter recipient's account number: ")
    recipient = bank.get_account_by_number(recipient_num)
    if recipient:
        amount = float(input("Enter the transfer amount: "))
        account.transfer(amount, recipient)
        account.update_account()
        recipient.update_account()
    else:
        print("Recipient account not found!")


def main():
    create_tables()

    while True:
        print("\n==== Bank Management System ====")
        print("1. Create Account")
        print("2. User Login")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            open_account()

        elif choice == "2":
            user = login()

            if isinstance(user, Admin):
                while True:
                    print("\n==== Admin Panel ====")
                    print("1. Account Statistics")
                    print("2. Logout")
                    option = input("Enter your choice: ")

                    if option == "1":
                        user.get_account_statistics()

                    elif option == "2":
                        break

                    else:
                        print("Invalid option!")

            elif isinstance(user, User):
                account_num = input("Enter your account number: ")
                account = bank.get_account_by_number(account_num)

                if account:
                    while True:
                        print("\n==== Account Management ====")
                        print("1. Deposit")
                        print("2. Withdraw")
                        print("3. Check Balance")
                        print("4. Transfer")
                        print("5. Change Password")
                        print("6. Logout")
                        option = input("Enter your choice: ")

                        if option == "1":
                            deposit(account)

                        elif option == "2":
                            withdraw(account)

                        elif option == "3":
                            check_balance(account)

                        elif option == "4":
                            transfer(account)

                        elif option == "5":
                            new_password = input("Enter new password: ")
                            user.change_password(new_password)

                        elif option == "6":
                            break

                        else:
                            print("Invalid option!")

                else:
                    print("Account not found!")

        elif choice == "3":
            print("Thank you for using the Bank Management System!")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    bank = Bank()
    admin_user = Admin("admin", "admin")
    user = User("user", "user")
    main()

# Close database connection
conn.close()
```
